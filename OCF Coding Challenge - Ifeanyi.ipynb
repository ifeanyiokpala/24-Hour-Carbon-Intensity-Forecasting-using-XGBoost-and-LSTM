### *XGBoost*

```python
# Adding new Features...

# Define the number of lags

lags = [1, 2, 4]

# Create lag features for relevant columns (carbon_intensity_actual, Solar, Wind Offshore, Wind Onshore)
for lag in lags:
    df[f'carbon_intensity_actual_lag_{lag}'] = df['Carbon_Intensity'].shift(lag)
    df[f'Solar_lag_{lag}'] = df['Solar'].shift(lag)
    df[f'Wind_Offshore_lag_{lag}'] = df['Wind_Offshore'].shift(lag)
    df[f'Wind_Onshore_lag_{lag}'] = df['Wind_Onshore'].shift(lag)

# Drop rows with missing values generated by shifting (at the beginning of the dataset)
df = df.dropna()
```

```python
df.info()
```

```python
# Using TimeSeriesSplit for Cross Validation Check 

from sklearn.model_selection import TimeSeriesSplit

tss = TimeSeriesSplit(n_splits=5, test_size=24*2, gap=24)
df = df.sort_index()


fold = 0
preds = []
scores = []
for train_idx, val_idx in tss.split(df):
    train = df.iloc[train_idx]
    test = df.iloc[val_idx]


    features_column = df.drop(columns=['Carbon_Intensity']).columns
    target_column = ['Carbon_Intensity']

    X_train = train[features_column]
    y_train = train[target_column]

    X_test = test[features_column]
    y_test = test[target_column]

    reg = xgb.XGBRegressor(base_score=0.5, booster='gbtree',    
                           n_estimators=1000,
                           early_stopping_rounds=50,
                           objective='reg:linear',
                           max_depth=3,
                           learning_rate=0.01)
    reg.fit(X_train, y_train,
            eval_set=[(X_train, y_train), (X_test, y_test)],
            verbose= False)

    y_pred = reg.predict(X_test)
    preds.append(y_pred)
    score = np.sqrt(mean_squared_error(y_test, y_pred))
    scores.append(score)
```

```python
print(f'Score across folds {np.mean(scores):0.4f}')
print(f'Fold scores:{scores}')
```

### *LSTM Model*

```python
# Deep Learning Modelling

# Addtional Import
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2



X = df.drop(columns=['Carbon_Intensity'])
y = df['Carbon_Intensity']

train_size = int(len(df)- (48*2)) 
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]


scaler = StandardScaler()

# Fit on training data and transform both train and test sets
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Reshape X data for LSTM model (samples, time steps, features)
# We will use a window of past 48 time steps to predict the next value
time_steps = 48

def create_sequences(X, y, time_steps):

    """
    windowing:
    Input Features = [X1, X2, …, X12], [X2, X3, …, X13], …, [X3, X4, …, X191]
    Labels = [X13], [X14], …, [X192]
    """
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        Xs.append(X[i:(i + time_steps)])
        ys.append(y[i + time_steps])
    return np.array(Xs), np.array(ys)

X_train_lstm, y_train_lstm = create_sequences(X_train_scaled, y_train, time_steps)
X_test_lstm, y_test_lstm = create_sequences(X_test_scaled, y_test, time_steps)


# Model Architecture

model = Sequential()
model.add(LSTM(units=32, return_sequences=True, kernel_regularizer=l2(0.01), input_shape=(X_train_lstm.shape[1], X_train_lstm.shape[2])))
model.add(Dropout(0.5))
model.add(LSTM(units=32, return_sequences=False, kernel_regularizer=l2(0.01)))
model.add(Dropout(0.5))
model.add(Dense(units=1))


model.compile(optimizer=Adam(learning_rate=0.0008), loss='mean_squared_error')

history = model.fit(X_train_lstm, y_train_lstm, epochs=50, batch_size=32, validation_data=(X_test_lstm, y_test_lstm))
```

```python
def loss_plot(history):
    
    """
    This function will be used to plot the loss
    between the validation and training performance
    from the history.
    
    *The history loss information
    
    """
    
    # Get mse from history.
    mse = history.history['loss']
    val_mse = history.history['val_loss']

    
    # Get number of epochs.
    epochs = range(len(history.epoch)) 


    # Plot mse loss.
    plt.figure(dpi=100, figsize = (20,10))
    plt.subplot(2,2,1)
    plt.grid(axis='y', linestyle='--', alpha= 0.8)
    plt.plot(epochs, mse)
    plt.plot(epochs, val_mse)
    plt.xlabel('Number of Epochs')
    plt.ylabel('Loss')
    plt.title('MSE Performance')
    plt.legend(['Training Loss', 'Validation Loss'])
    
```

```python
loss_plot(history)
```

```python
y_pred_lstm = model.predict(X_test_lstm)

model_plot(df,df2=y_pred_lstm,model_name= 'LSTM Model',zoom=17500, title= 'LSTM Carbon Intensity Forecast vs Actual',legend2= 'Predicted')
```
